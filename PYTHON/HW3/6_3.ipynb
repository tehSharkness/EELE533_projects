```python
from numpy import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.arange(0, 2*pi, pi/2**16)

AF = sin(3*pi/2*cos(theta) + 3*pi/4)/sin(pi/2 * cos(theta) + pi/4)
```

```python
plt.plot(theta/pi, AF)
```

```python
print('{} at theta={}'.format(np.amax(AF), theta[np.argmax(AF)]/pi))
```
